#### Basic Graph Terminology
A graph consists of **nodes** and **edges**. A node/vertex is a point on the graph. An edge is a line connecting two nodes. 



A **path** is an ordered sequences of edges that leads from one vertex to another. Its length is the number of edges in the sequence.


A graph is **connected** if there is a path between any two nodes.

A **component** is all connected parts of the graph.


Two nodes are **neighbors/adjacent** if an edge connects them.

######  Two Types of Graphs
The first type of graph is a **directed** graph which means that all edges can only be traversed in one direction.


A graph is **undirected** if all edges can be traversed in both directions.


###### More terms
A **cycle** is a path if the start node is the same as the end node. 

A graph is **cyclic** if it contains at least one cycle.

If a graph is **weighted**, then each edge is assigned a **weight** - often interpereted as edge lengths.
