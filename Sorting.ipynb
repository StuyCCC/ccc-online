### Sorts
Sorting algorithms are algorithms used to put arrays into order. There are tons of sorting algorithms, all with different use cases and time complexities. Some may have the same time complexity, but perform wildly differently depending on the input data's properties and other variables. We'll be explaining a few of the most common and useful sorting algorithms here.

#### Merge Sort
Merge sort works by splitting the input array into "chunks" recursively until each "chunk" is 2 integers long, this process is often called a "divide and conquer" approach. It then sorts each of these chunks, by checking which number is greater. Finally, it recursively merges the chunks, preserving the order, until the input array has been reassembled.

Merge Sort runs in `O(nlogn)` on average




```python
#You don't need to memorize this code, only the algorithm behind it.
def merge_lists(a, b):
    new_list = []
    latest_a = 0
    latest_b = 0
    while latest_a < len(a) and latest_b < len(b):
        if a[latest_a] < b[latest_b]:
            new_list.append(a[latest_a])
            latest_a += 1
        else:
            new_list.append(b[latest_b])
            latest_b += 1
    new_list += a[latest_a:] + b[latest_b:]
    return new_list


def merge_sort(g):
    if len(g) <= 1:
        return g
    else:
        g1 = g[:(len(g) // 2)]
        g2 = g[(len(g) // 2):]
        return merge_lists(merge_sort(g1), merge_sort(g2))
```

#### Quick Sort
Quick Sort is another divide and conquer sorting algorithm. Quick Sort uses a pivot to split the array, placing all elements that are greater than the pivot to the right, and smaller to the left. Then, each sub-array has Quick Sort run recursively on it.

While Quick Sort has the same upper bound as Merge Sort, the lower bound of the time complexity can be lower. We still say that it runs in `O(nlogn)`.

```python
#Quick Sort implementation, if you're interested. You don't need to know this.
def partition(arr, low, high):
    i = (low-1)
    pivot = arr[high] 
    for j in range(low, high):
        if arr[j] <= pivot:
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)
  
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
```

#### Binary Search
When you need to find an element in an array, you can either iterate over the array until you find the element, or you can employ the use of a search algorithm. One of these search algorithms is binary search. Binary search operates on sorted arrays. It works by checking the value at the middle of the array. If the middle is greater than the value you're searching for, you re-run binary search on the first half of the array. If it's less than the value you're searching for, do the same, but run it on the second half of the array. 

The runtime of a binary search is `O(logn)`.

```python
#A Recursive Binary Search Implementation
def binary_search(array, element, start, end):
    if start > end:
        return -1
    mid = (start + end) // 2
    if element == array[mid]:
        return mid
    if element < array[mid]:
        return binary_search(array, element, start, mid-1)
    else:
        return binary_search(array, element, mid+1, end)
```
